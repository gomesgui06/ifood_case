### Format and create columns

```python
df.shape
```

```python
df_trusted = df.copy()
df_trusted.shape
```

```python
df_trusted['income'] = df.income.fillna(df.income.describe()['50%'])
df_trusted.shape
```

```python
df_trusted.isna().sum()
```

```python
df_trusted['date_customer'] = pd.to_datetime(df_trusted['dt_customer'])
df_trusted.head()
```

```python
df_trusted['age'] = datetime.today().year - df_trusted['year_birth']
```

```python
df.info()
```

```python
df_trusted.marital_status.unique()
```

```python
df_trusted.education.unique()
```

```python
df_trusted['current_date'] = pd.Timestamp('today').strftime("%Y-%m-%d")
df_trusted['current_date'] = pd.to_datetime(df_trusted.current_date)
```

```python
df_trusted['client_days'] = (df_trusted.current_date - df_trusted.date_customer).dt.days
```
